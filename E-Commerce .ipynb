####  *Concept used*
____

y = m1x1+m2x2+c

y - dependent variable 

x1,x2,x3 as independent variable

i have missing values in x1

dependent var = x1

independent var = x2 and x3

we are trying to predict x1 using x2 and x3 

traing set = where we have values for x1

test set = where we dont have values for x1

```python
# Using Z-Score Method

scalled_df['Tenure'] = (scalled_df['Tenure'] - scalled_df['Tenure'].mean()) / scalled_df['Tenure'].std()
scalled_df['WarehouseToHome'] = (scalled_df['WarehouseToHome'] - scalled_df['WarehouseToHome'].mean()) / scalled_df['WarehouseToHome'].std()
scalled_df['OrderAmountHikeFromlastYear'] = (scalled_df['OrderAmountHikeFromlastYear'] - scalled_df['OrderAmountHikeFromlastYear'].mean()) / scalled_df['OrderAmountHikeFromlastYear'].std()
scalled_df['CashbackAmount'] = (scalled_df['CashbackAmount'] - scalled_df['CashbackAmount'].mean()) / scalled_df['CashbackAmount'].std()
scalled_df['CouponUsed'] = (scalled_df['CouponUsed'] - scalled_df['CouponUsed'].mean()) / scalled_df['CouponUsed'].std()
```

# Using Min-Max Method
